```python
%load_ext autoreload
%autoreload 2

import os
from utils.llm import LLM_USE_CACHE
from nodes.keyword_extractor_graph import run_extractor
from rexpand_pyutils_file import read_file, write_file

print('DEEPSEEK_API_KEY set:', bool(os.getenv('DEEPSEEK_API_KEY')))
print('LLM model:', os.getenv('LLM_MODEL', 'deepseek-chat'))
print('LLM use cache:', LLM_USE_CACHE)
```

```python
# Load document
file_path = 'examples/job_description.md'
document_text = read_file(file_path)
filename = file_path.split('/')[-1].split('.')[0]
```

```python
# Run full extractor (Phase 1 + Phase 2) to build keyword datapoints
state = run_extractor(document_text, document_type="jd")
write_file(f"./output/{filename}_phase_1_state.json", state.model_dump())
state
```

```python
from matcher.matcher import run_matcher

jd_text = read_file('examples/job_description.md')
resume_text = read_file('examples/resume.md')

jd_state = run_extractor(jd_text, document_type='jd')
resume_state = run_extractor(resume_text, document_type='resume')

match = run_matcher(jd_state.datapoints.skills, resume_state.datapoints.skills)
print(f"Precision: {match.precision}")
print(f"Recall: {match.recall}")
print(f"F1: {match.f1}")
print(f"Is match: {match.is_match}")
print('Matched:')
for pair in match.matched:
    print(f"  - {pair.jd_skill.name} <-> {pair.resume_skill.name} ({pair.similarity})")
print('Missing from resume:')
for skill in match.missing_from_resume:
    print(f"  - {skill.name}")
```
